### Ejercicio 4 
El ESTUDIANTE 2, creando un “branch” llamado ejerPrograma, generará un programa que muestre los números primos del 0-100.

```python
# Código de solución estudiante 2
# del 0 al 100, entonces realizamo sun ciclo for, del 0 al 101 ya que no toma en cuenta el último entonces sería 101-1=100
for n in range(0, 101):
    primo = True # Asumimos que es primo desde el principio y comenzamos una sentencia if
    if n < 2:
        primo = False
        # Sabemos que los número primos son mayores de 1, entonces si es menor no es un número primo
    else:
        for i in range(2, n): # Entonces comenzamos desde el 2 hasta n-1
            if n % i == 0:  # Corroboramos cada iteracion si n  es divisible entre i
                primo = False # Si esto sucede se asigna falso
                break  # Rompemos el ciclo ya que si se encuentra un número sabremos que no es primo y continuamos con los demás
    if primo:
        print(n, end=' ')
    # Entonces, sí es primo lo imprimimos, se le agrega el end con espacio, para que lo imprima en una sola línea esto por visualización.
```
